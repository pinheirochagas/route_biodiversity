```python
import numpy as np
import matplotlib.pyplot as plt

# Months
months = np.arange(1, 13)

# Species 1: High base rate, peak in March
mu_1 = 3  # March
sigma_1 = 1  # Standard deviation
base_rate_1 = 1000  # Base observation rate
peak_1 = 8000  # Peak observation rate
observations_1 = base_rate_1 + (peak_1 - base_rate_1) * np.exp(-(months - mu_1)**2 / (2 * sigma_1**2))

# Species 2: Low base rate, peak in October
mu_2 = 10  # October
sigma_2 = 0.5  # Standard deviation
peak_2 = 12  # Peak observation rate
observations_2 = peak_2 * np.exp(-(months - mu_2)**2 / (2 * sigma_2**2))

# Visualization
plt.figure(figsize=(10, 6))
#plt.plot(months, observations_1, label='Species 1', marker='o')
plt.plot(months, observations_2, label='Species 2', marker='o')
plt.title('Simulated Monthly Observations for Two Species')
plt.xlabel('Month')
plt.ylabel('Observations')
plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ... [Previous code for observations_1 and observations_2]

# Calculate mean and std dev for each species
mean_1, std_dev_1 = np.mean(observations_1), np.std(observations_1)
mean_2, std_dev_2 = np.mean(observations_2), np.std(observations_2)

# Calculate z-scores
z_scores_1 = (observations_1 - mean_1) / std_dev_1
z_scores_2 = (observations_2 - mean_2) / std_dev_2

# Visualization of z-scores
plt.figure(figsize=(10, 6))
plt.plot(months, z_scores_1, label='Species 1', marker='o')
plt.plot(months, z_scores_2, label='Species 2', marker='o')
plt.title('Z-Scores of Monthly Observations for Two Species')
plt.xlabel('Month')
plt.ylabel('Z-Score')
plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.axhline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
